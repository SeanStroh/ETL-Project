```python
from sqlalchemy import create_engine
from sqlalchemy import Column, Integer, String, Float
from sqlalchemy.ext.declarative import declarative_base
Base = declarative_base()
import pandas as pd
import psycopg2
```

```python
Base.metadata.tables
```

```python
class airplaines(Base):
  __tablename__ = 'plane_datass'
  id = Column(Integer, primary_key=True)
  Passanger_Enplanement = Column(Integer)
  Aircraftzz_Hours_Flown = Column(Integer)
  Hull_Losses =Column(Integer)
  Hull_Losses_per_Million_Aircraft_Hours_Flown =Column(Integer)
  Major_Accidents =Column(Integer)
  Major_Accidents_per_Million_Aircraft_Hours_Flown =Column(Integer)
  Serious_Accidents =Column(Integer)
  Serious_Accidents_per_Million_Aircraft_Hours_Flown =Column(Integer)
  Injuries =Column(Integer)
  Injuries_per_Million_Aircraft_Hours_Flown =Column(Integer)
  Damages =Column(Integer)
  Damage_per_Million_Aircraft_Hours_Flown = Column(Integer)
  Passenger_Fatalities = Column(Integer)
  Serious_Passenger_Injuries = Column(Integer)
```
